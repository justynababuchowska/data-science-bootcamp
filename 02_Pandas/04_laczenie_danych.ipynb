```python
import pandas as pd
```

Wczytanie danych

```python
def fetch_finantial_data(company='AMZN'):
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

apple = fetch_finantial_data('AAPL')
amazon = fetch_finantial_data('AMZN')
google = fetch_finantial_data('GOOGL')
uber = fetch_finantial_data('UBER')
```

```python
apple.head()
```

```python
google.head()
```

Prepocessing

```python
apple.columns = ['apple_' + name.lower() for name in apple.columns]
```

```python
apple.columns
```

```python
amazon.columns = ['amazon_'+ name.lower() for name in amazon.columns]
google.columns = ['google_'+ name.lower() for name in google.columns]
uber.columns= ['uber_'+ name.lower() for name in uber.columns]
```

Konkatenacja danych = połączenie danych

```python
#łączymy dane po obiektach zawietających dataFrame i podajemy względem czego mają być połączone czyli axis=1 (kolumny przyklejamy do siebie obok siebie), axis=0 kolumny przyklejane w pionie pod sobą
df = pd.concat(objs=[apple, amazon, google, uber], axis=1)
df
```

```python
#ustawiamy opcje wyświetlania, aby były 2miejsca po przecinku pd.set_options(co chcemy zmienić, na co zmieniamy)
pd.set_option('display.float_format',lambda x: f'{x:.2f}')
df.describe().T
```

```python
#korelacja .corr() - określa, w jaki sposób dwie zmienne są związane ze sobą. Jeśli dwie zmienne są dodatnio skorelowane, oznacza to, że wzrost jednej zmiennej jest powiązany z wzrostem drugiej zmiennej.
#Natomiast jeśli dwie zmienne są ujemnie skorelowane, oznacza to, że wzrost jednej zmiennej jest powiązany ze spadkiem drugiej zmiennej.
#Korelacja bliska zeru oznacza brak związku liniowego między zmiennymi.
df.corr()
```

```python
#liczymy współczynnik korelacji tylko dla wartości kolumn w close
#szukamy nazw kolumn, które mają w nazwie 'close'

closes = [name for name in df.columns if name.endswith('close')]
closes
```

```python
#wycinamy dane dla kolumn closes
df[closes]
```

```python
#możemy policzyć korelacje między sobą
df[closes].corr()
```

```python
closes = [name for name in df.columns if name.endswith('close') or name.endswith('volume')]
closes
```

```python
df[closes].corr()
```
