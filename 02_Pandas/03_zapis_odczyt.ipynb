```python
import numpy as np
import pandas as pd
```

Załadowanie danych

```python
def fetch_finantial_data(company = 'AMZN'):
    import pandas_datareader.data as web
    return web.DataReader(name = company, data_source= 'stooq')

df = fetch_finantial_data('META')
df.head()
```

Zapis/Odczyt danych z pliku CSV

```python
#zapis powyższego pliku do CSV - w nawiasie podajemy jaka będzie nazwa pliku
df.to_csv('fb.csv')
```

```python
#wycinamy dane z november 2019
df_nov = df[(df.index.month == 11)&(df.index.year == 2019)]
df_nov
```

```python
#zapisujemy powyższe dane do pliku CSV
df_nov.to_csv('fb_nov.csv')
```

```python
#wczytujemy dane z pliku CSV
new_df = pd.read_csv('fb_nov.csv')
new_df
```

```python
#zapisujemy plik do Excel analogicznie jak CSV
df_nov.to_excel('fb_nov.xlsx')
```

```python
#podczas odczytu nasz index (date) zamienił się na kolumnę i jeśli chcemy to zmienić po przecinku index_col to ustawiamy
new_df = pd.read_excel('fb_nov.xlsx', index_col = 0)
new_df
```
