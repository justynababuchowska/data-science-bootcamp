Obliczanie nowych kolumn

```python
df = fetch_financial_data('UBER')
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.head(3)
```

```python
#obliczamy w nowej kolumnie średnią cenę z open oraz close
df['Average'] = (df.Open + df.Close) / 2.0
df.head(3)
```

```python
#sortujemy dane od najstarszych i przypisujemy do df aby się zapisało
df= df.sort_index()
```

```python
df.head()
```

```python
#liczymy w nowej kolumnie dzienną zmianę procentową
#dla 1dnia nie znamy danych zamknięcia z poprzedniego dnia, więc przesuwamy w nowej kolumnie dane o1miejsce - różnica danego dnia z otwarcia i zamknięcia poprzedniego dnia
#funkcja .shift() przesuwa nam dane o tyle miejsc ile wpiszemy w nawiasie

df['Close_shift'] = df.Close.shift(1)
df
```

```python
df['Daily_change'] = df.Close / df.Close_shift - 1
df.head()
```

```python
df['do usuniecia'] = np.nan
df.head()
```

```python
#usunięcie kolumny
del df['do usuniecia']
df.head()
```
