```python
import pandas as pd
import numpy as np
```

Wczytywanie danych i preprocessing

```python
url = ('https://storage.googleapis.com/esmartdata-courses-files/ds-bootcamp/online_retail.xlsx')
retail_raw = pd.read_excel(url)
retail_raw.head()
```

```python
retail = retail_raw.copy()
retail.info()
```

```python
retail.describe()
```

```python
retail.isnull().sum()
```

```python
#wycinamy wiersze z wartościami null
retail = retail.dropna()
retail.isnull().sum()
```

```python
#usówamy z kolumny qantity wartości ujemne - jako nieprawidłowe
retail = retail[retail.Quantity>=0]
```

```python
retail.info()
```

```python
#zmieniamy typ customerId na int a następnie na str (w info wyświetli się jako object)
retail.CustomerID = retail.CustomerID.apply(lambda x: str(int(x)))
retail.info()
```

```python
retail.head()
```

```python
#select CustomerID, count(*) from retail group by CustomerID
retail.groupby('CustomerID').count() #zle, bo mamy całą tabelę
```

```python
#size() zliczy nam tylko parametr CustomerID - częstość występowania
retail.groupby('CustomerID').size()
```

```python
#select CustomerID, avg(Revenue), count(*) from retail group by CustomerID
#na początku musimy utworzyć kolumne z Revenue
retail.head(1)
```

```python
#mnożymy quantity*unitprice aby dostać pojedyńczą tranzakcję
retail['Revenue'] = retail.Quantity * retail.UnitPrice
retail.head(3)
```

```python
#jeśli chcemy użyć 2funkcji zliczających size() oraz avg musimy użyć funkcji agregującej aggregate() i za pomocą słownika mapujemy kolumny na odpowiednie funkcje
retail.groupby('CustomerID').aggregate({'Revenue': np.mean, 'CustomerID': np.size})
```

```python
#zmieniamy nazwy kolumn - też mapując odpowiednie kolumny za pomocą słownika
retail.groupby('CustomerID').aggregate({'Revenue': np.mean, 'CustomerID': np.size}).rename(columns= {'Revenue': 'RevenueAverage', 'CustomerID': 'CustomerIDCount'})
```

```python
#select InvoiceDateDay, sum(Revenue) from retail group by InvoiceDateDay
#tworzymy na początku kolumne InvoiceDateDay
#parametrem dt możemy przywołać metody, które opedują na date time i przywołać day - tylko jeśli mamy wartości typu date
retail['InvoiceDateDay'] = retail.InvoiceDate.dt.day
retail.head(1)
```

```python
#grupujemy kolumny i sumujemy revenue za pomocą funkcji agregującej
results = retail.groupby('InvoiceDateDay').aggregate({'Revenue': np.sum}).rename(columns = {'Revenue':'Revenue_sum'})
results
```

```python
#wykres słupkowy
import seaborn as sns
sns.set()

_=results.plot(kind='bar')
```
